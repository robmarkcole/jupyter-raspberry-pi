We can use the picamera library rather than faff with openCV

```python
import picamera
camera = picamera.PiCamera() 
camera.vflip = True
```

```python
file_name = 'img.jpg'
camera.capture(file_name)
```

```python
img = plt.imread(file_name)
plt.imshow(img);
```
